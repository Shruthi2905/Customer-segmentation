# K-means clustering is a popular unsupervised machine learning algorithm used for clustering similar data points into a specified number of clusters. It's a simple yet effective algorithm that's widely used for data exploration, segmentation, and pattern recognition.

1: Decide the Number of Cluster

2: Initialise Centroids , Take Random Values

3: Find Minimum Distance using Euclidean Distance Formula and Assign the Cluster

4: Find New Centroid by averaging out the points

5: Goto Step 3 to again find/assign clusters

6: Stop when the centroids stop moving

Elbow Method: 

WCSS - Within-Cluster Sum of Squares: WCSS value indicates that the data points within each cluster are how much closer to their respective centroids, resulting in tighter and more compact clusters.

    Formula - Dis12 + Dis22 + Dis32 ......
    WCSS is Max at 1 Cluster and Min at N Clusters (N>Total Number of DataPoints)

Elbow Point : In the Elbow Point Graph, find the point where the rate of variance decreases sharply levels off
    
    

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Data Collection & Analysis

```python
# loading the data from csv file to a Pandas DataFrame
customer_data = pd.read_csv('Mall_Customers.csv')
```

```python
# getting some informations about the dataset
customer_data.info()
```

```python
# checking for missing values
customer_data.isnull().sum()
```

Choosing the Annual Income Column & Spending Score column

```python
X = customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

Choosing the number of clusters

WCSS  ->  Within Clusters Sum of Squares

```python
# finding wcss value for different number of clusters

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

n_clusters=i: This parameter specifies the number of clusters to form. The value of i is typically determined based on some criteria, such as the "elbow method".

init='k-means++': This parameter specifies the method for initialization of centroids. 'k-means++' is a smart initialization method that selects initial cluster centroids in a way that speeds up convergence.
        
random_state=42: This parameter sets the random seed for the initialization of centroids, ensuring reproducibility of results.

```python
# plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clusters = 5

Training the k-Means Clustering Model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)

print(Y)
```

5 Clusters -  0, 1, 2, 3, 4

Visualizing all the Clusters

```python
# plotting all the clusters and their Centroids

plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
